```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
```

```python
y = np.sin(x)
```

```python
z = np.cos(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.xlabel("angle")
plt.ylabel("value")
plt.title("cos wave")
plt.plot(x,z)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ["English", "French", "Hindi"]
people = [100, 50, 150]
ax.bar(languages,people)
plt.xlabel("languages")
plt.ylabel("No. of People")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ["English", "French", "Hindi"]
people = [100, 50, 150]
ax.pie(people,labels=languages)
plt.show()
```
